## –ü—Ä–µ–¥–æ–±—Ä–∞–±–æ—Ç–∫–∞ –∏ –ø–æ–¥–≥–æ—Ç–æ–≤–∫–∞ –¥–∞–Ω–Ω—ã—Ö

–ü—Ä–∏–≤–µ–¥—ë–º –¥–∞–Ω–Ω—ã–µ –æ –≤—Ä–µ–º–µ–Ω–∏ –≤ –Ω—É–∂–Ω—ã–π —Ñ–æ—Ä–º–∞—Ç:

–í—ã—á–∏—Å–ª–∏–º –≤–æ–∑—Ä–∞—Å—Ç, –≤ –∫–æ—Ç–æ—Ä–æ–º –∞–≤—Ç–æ—Ä—ã —Å–æ–∑–¥–∞–ª–∏ —Å–≤–æ–∏ –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏—è:

–í—ã–¥–µ–ª–∏–º –≥–æ–¥ –ø—Ä–æ—á—Ç–µ–Ω–∏—è:

–í—ã–¥–µ–ª–∏–º –≤–µ–∫ –Ω–∞–ø–∏—Å–∞–Ω–∏—è –∫–Ω–∏–≥–∏: 

```python
def find_century (year_writing):
    """
    –í–æ–∑–≤—Ä–∞—â–∞–µ—Ç –≤–µ–∫ –ø–æ –∑–Ω–∞—á–µ–Ω–∏—é –≥–æ–¥–∞ –Ω–∞–ø–∏—Å–∞–Ω–∏—è –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏—è
    
    """
    if year_writing <= 1400:
        return '14-–π –≤–µ–∫'
    elif year_writing <= 1500:
        return '15-–π –≤–µ–∫'
    elif year_writing <= 1600:
        return '16-–π –≤–µ–∫'
    elif year_writing <= 1700:
        return '17-–π –≤–µ–∫'
    elif year_writing <= 1800:
        return '18-–π –≤–µ–∫'
    elif year_writing <= 1900:
        return '19-–π –≤–µ–∫'
    elif year_writing <= 2000:
        return '20-–π –≤–µ–∫'
    elif year_writing <= 2100:
        return '21-–π –≤–µ–∫'
    else:
        return '–Ω–µ–∏–∑–≤–µ—Å—Ç–Ω–æ'
```

```python
df['century_writing'] = df['year_writing'].apply(find_century) 
```

–í—ã–¥–µ–ª–∏–º –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ –≤–æ–∑—Ä–∞—Å—Ç–æ–≤ –∞–≤—Ç–æ—Ä–æ–≤ –Ω–∞ –º–æ–º–µ–Ω—Ç –Ω–∞–ø–∏—Å–∞–Ω–∏—è –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–π:

```python
def find_age_category (age_writing):
    """
    –í–æ–∑–≤—Ä–∞—â–∞–µ—Ç –∫–∞—Ç–µ–≥–æ—Ä–∏—é –≤–æ–∑—Ä–∞—Å—Ç–∞ –ø–æ –∑–Ω–∞—á–µ–Ω–∏—é –≤–æ–∑—Ä–∞—Å—Ç–∞ –∞–≤—Ç–æ—Ä–∞ –Ω–∞ –º–æ–º–µ–Ω—Ç –Ω–∞–ø–∏—Å–∞–Ω–∏—è –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏—è
    
    """
    if age_writing < 30:
        return '–¥–æ 30 –ª–µ—Ç'
    elif 30 <= age_writing < 40:
        return '–æ—Ç 30 –¥–æ 40 –ª–µ—Ç'
    elif 40 <= age_writing < 50:
        return '–æ—Ç 40 –¥–æ 50 –ª–µ—Ç'
    elif 50 <= age_writing < 60:
        return '–æ—Ç 50 –¥–æ 60 –ª–µ—Ç'
    elif 60 <= age_writing < 70:
        return '–æ—Ç 60 –¥–æ 70 –ª–µ—Ç'
    elif age_writing >= 70:
        return '–æ—Ç 70 –ª–µ—Ç –∏ —Å—Ç–∞—Ä—à–µ'
```

```python
df['age_writing_category'] = df['age_writing'].apply(find_age_category)
```

–ó–∞–¥–∞–¥–∏–º –±–æ–ª–µ–µ –æ–±–æ–±—â—ë–Ω–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–æ —Å—Ç—Ä–∞–Ω–∞–º:

```python
def find_country_type (author_country):
    """
    –í–æ–∑–≤—Ä–∞—â–∞–µ—Ç –ø—Ä–∏–Ω–∞–¥–ª–µ–∂–Ω–æ—Å—Ç—å –∞–≤—Ç–æ—Ä–∞ –∫ —Å—Ç—Ä–∞–Ω–µ –ø–æ –±–æ–ª–µ–µ –æ–±–æ–±—â—ë–Ω–Ω—ã–º –ø–∞—Ä–∞–º–µ—Ç—Ä–∞–º 
    (–Ω–∞–ø—Ä–∏–º–µ—Ä, –§—Ä–∞–Ω—Ü–∏—è, –ò—Å–ø–∞–Ω–∏—è –∏ –ø—Ä–æ—á–∏–µ –µ–≤—Ä–æ–ø–µ–π—Å–∫–∏–µ —Å—Ç—Ä–∞–Ω—ã –æ–±—ä–µ–¥–∏–Ω–µ–Ω—ã –≤ –æ–¥–Ω–æ –Ω–∞–∏–º–µ–Ω–æ–≤–∞–Ω–∏–µ - –ï–≤—Ä–æ–ø–∞)
    
    """
    if author_country == '–°–®–ê':
        return '–°–®–ê'
    elif author_country == '–†–æ—Å—Å–∏—è' or author_country == '–°–°–°–†':
        return '–†–æ—Å—Å–∏—è'
    elif author_country == '–í–µ–ª–∏–∫–æ–±—Ä–∏—Ç–∞–Ω–∏—è':
        return '–í–µ–ª–∏–∫–æ–±—Ä–∏—Ç–∞–Ω–∏—è'
    elif author_country == '–Ø–ø–æ–Ω–∏—è':
        return '–Ø–ø–æ–Ω–∏—è'
    elif author_country in ['–ò—Å–ø–∞–Ω–∏—è', '–§—Ä–∞–Ω—Ü–∏—è', '–ê–≤—Å—Ç—Ä–∏—è', '–ü–æ–ª—å—à–∞', '–†—É–º—ã–Ω–∏—è', '–ì–µ—Ä–º–∞–Ω—Å–∫–∞—è –∏–º–ø–µ—Ä–∏—è, –®–≤–µ–π—Ü–∞—Ä–∏—è', 
                            '–§–ª–æ—Ä–µ–Ω—Ç–∏–π—Å–∫–∞—è —Ä–µ—Å–ø—É–±–ª–∏–∫–∞', '–°–≤—è—â–µ–Ω–Ω–∞—è –†–∏–º—Å–∫–∞—è –∏–º–ø–µ—Ä–∏—è, –ì–µ—Ä–º–∞–Ω—Å–∫–∏–π —Å–æ—é–∑']:
        return '–ï–≤—Ä–æ–ø–∞'
    elif author_country in ['–¢—É—Ä—Ü–∏—è', '–ë–µ–ª–∞—Ä—É—Å—å', '–ö–æ–ª—É–º–±–∏—è', '–ê—Ä–º–µ–Ω–∏—è', '–†–æ—Å—Å–∏–π—Å–∫–∞—è –∏–º–ø–µ—Ä–∏—è, –°–®–ê', '–ò–∑—Ä–∞–∏–ª—å, –°–®–ê']:
        return '–¥—Ä—É–≥–æ–µ'
```

```python
df['author_country_type'] = df['author_country'].apply(find_country_type)
```

–í—ã—á–∏—Å–ª–∏–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –¥–Ω–µ–π –æ—Ç –ø—Ä–æ—á—Ç–µ–Ω–∏—è –æ–¥–Ω–æ–π –∫–Ω–∏–≥–∏ –¥–æ –ø—Ä–æ—á—Ç–µ–Ω–∏—è –¥—Ä—É–≥–æ–π:

```python
df['days_reading_cnt'] = df['date_reading'] - df['date_reading'].shift(1)
```
