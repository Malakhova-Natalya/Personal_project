–î–ª—è —Ç–æ–≥–æ, —á—Ç–æ–±—ã –±—ã–ª–æ —É–¥–æ–±–Ω–æ —Ä–∞—Å—Å–º–∞—Ç—Ä–∏–≤–∞—Ç—å –±–æ–ª—å—à–∏–µ —á–∏—Å–ª–∞, —Å–¥–µ–ª–∞–µ–º —Ä–∞–∑–¥–µ–ª–µ–Ω–∏–µ –ø—Ä–æ–±–µ–ª–∞–º–∏:

```python
class _IntArrayFormatter(pd.io.formats.format.GenericArrayFormatter):

    def _format_strings(self):
        formatter = self.formatter or (lambda x: ' {:,}'.format(x).replace(',', ' '))
        fmt_values = [formatter(x) for x in self.values]
        return fmt_values

pd.io.formats.format.IntArrayFormatter = _IntArrayFormatter
```

–î–ª—è –æ–∑–Ω–∞–∫–æ–º–ª–µ–Ω–∏—è –∏ –ø—Ä–æ–≤–µ—Ä–∫–∏ –¥–∞–Ω–Ω—ã—Ö –±—É–¥–µ–º –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å —Å–æ–±—Å—Ç–≤–µ–Ω–Ω—É—é —Ñ—É–Ω–∫—Ü–∏—é:

```python
def my_check_function (dataset):
    '''–§—É–Ω–∫—Ü–∏—è –¥–ª—è –æ–∑–Ω–∞–∫–æ–º–ª–µ–Ω–∏—è –∏ –ø—Ä–æ–≤–µ—Ä–∫–∏ –¥–∞–Ω—ã—Ö'''
    print('–†–∞–∑–º–µ—Ä –¥–∞—Ç–∞—Å–µ—Ç–∞ (–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—Ç—Ä–æ–∫, —Å—Ç–æ–ª–±—Ü–æ–≤):')
    print(dataset.shape)
    print('–ö–∞–∫ –≤—ã–≥–ª—è–¥—è—Ç —Å–ª—É—á–∞–π–Ω—ã–µ 5 —Å—Ç—Ä–æ–∫:')
    display(dataset.sample(5, random_state=42))
    print()
    print('–û–±—â–∞—è –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –æ –¥–∞–Ω–Ω—ã—Ö, –Ω–∞–∏–º–µ–Ω–æ–≤–∞–Ω–∏—è —Å—Ç–æ–ª–±—Ü–æ–≤, —Ç–∏–ø—ã –¥–∞–Ω–Ω—ã—Ö:')
    print(dataset.info())
    print()
    print('–ï—Å—Ç—å –ª–∏ –ø—Ä–æ–ø—É—Å–∫–∏, —Å–∫–æ–ª—å–∫–æ –∏—Ö:')
    print(dataset.isna().sum())
    print('–î–æ–ª—è –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π:')
    print(dataset.isna().mean())
    print()
    print('–ï—Å—Ç—å –ª–∏ —è–≤–Ω—ã–µ –¥—É–±–ª–∏–∫–∞—Ç—ã, —Å–∫–æ–ª—å–∫–æ –∏—Ö:')
    print(dataset.duplicated().sum())
    print('–î–æ–ª—è —è–≤–Ω—ã—Ö –¥—É–±–ª–∏–∫–∞—Ç–æ–≤:')
    print(dataset.duplicated().mean())
    print()
    print('–ü–æ–¥—Ä–æ–±–Ω–æ–µ –æ–ø–∏—Å–∞–Ω–∏–µ –¥–∞–Ω–Ω—ã—Ö:')
    display(dataset.describe())
```

–í–æ—Ç —É –∏—Å–ø–æ–ª–Ω–∏—Ç–µ–ª—è –ø–æ—è–≤–∏–ª—Å—è –∫–∞–Ω–∞–ª –Ω–∞ –≤–∏–¥–µ–æ–ø–ª–∞—Ç—Ñ–æ—Ä–º–µ. –ü–æ—è–≤–∏–ª–∏—Å—å –ø–µ—Ä–≤—ã–µ –≤–∏–¥–µ–æ, –ø–æ—è–≤–∏–ª–∏—Å—å –ø–µ—Ä–≤—ã–µ –ø—Ä–æ—Å–º–æ—Ç—Ä—ã. –ê —Å–∫–æ–ª—å–∫–æ —á–µ–ª–æ–≤–µ–∫ –ø–æ–¥–ø–∏—Å–∞–ª–æ—Å—å –Ω–∞ –∫–∞–Ω–∞–ª? –°–æ–ø–æ—Å—Ç–∞–≤–∏–º—ã –ª–∏ —É—Ä–æ–≤–Ω–∏ –ø—Ä–æ—Å–º–æ—Ç—Ä–æ–≤ –∏ –ø–æ–¥–ø–∏—Å—á–∏–∫–æ–≤? 

–†–∞—Å—Å—á–∏—Ç–∞–µ–º –ø—Ä–æ—Ü–µ–Ω—Ç –ø—Ä–∏–≤–ª–µ—á–µ–Ω–∏—è –∫–∞–∫ —Å–æ–æ—Ç–Ω–æ—à–µ–Ω–∏–µ —á–∏—Å–ª–∞ –ø–æ–¥–ø–∏—Å—á–∏–∫–æ–≤ –∏ –ø—Ä–æ—Å–º–æ—Ç—Ä–æ–≤. 

- ***–µ—Å–ª–∏ —ç—Ç–æ—Ç –ø–æ–∫–∞–∑–∞—Ç–µ–ª—å –Ω–µ–±–æ–ª—å—à–æ–π*** - –∑–Ω–∞—á–∏—Ç, –ø–æ–¥–ø–∏—Å—á–∏–∫–æ–≤ –º–∞–ª–æ, –∏ –º–∞–ª–æ –∫—Ç–æ –∏–∑ –ø–æ—Å–º–æ—Ç—Ä–µ–≤—à–∏—Ö –ø–æ–¥–ø–∏—Å–∞–Ω –Ω–∞ –∫–∞–Ω–∞–ª (–Ω–∞–ø—Ä–∏–º–µ—Ä –µ—Å–ª–∏ 1 —á–µ–ª–æ–≤–µ–∫ –∏–∑ 100 –ø–æ—Å–º–æ—Ç—Ä–µ–≤—à–∏—Ö –ø–æ–¥–ø–∏—Å–∞–ª—Å—è –Ω–∞ –∫–∞–Ω–∞–ª - –¥–∞–Ω–Ω–∞—è –º–µ—Ç—Ä–∏–∫–∞ –ø–æ–∫–∞–∂–µ—Ç 1%). –≠—Ç–æ –∑–Ω–∞—á–∏—Ç —É –∫–∞–Ω–∞–ª–∞ –Ω–µ—Ç —É—Å—Ç–æ—è–≤—à–µ–π—Å—è –≤–æ–≤–ª–µ—á—ë–Ω–Ω–æ–π –∞—É–¥–∏—Ç–æ—Ä–∏–∏. 


- ***–µ—Å–ª–∏ –∂–µ –ø–æ–∫–∞–∑–∞—Ç–µ–ª—å –±–æ–ª—å—à–æ–π*** (–Ω–∞–ø—Ä–∏–º–µ—Ä –∏–∑ 100 –ø–æ—Å–º–æ—Ç—Ä–µ–≤—à–∏—Ö –ø–æ–¥–ø–∏—Å–∞–ª–∏—Å—å 100 —á–µ–ª–æ–≤–µ–∫ - –º–µ—Ç—Ä–∏–∫–∞ –±—É–¥–µ—Ç 100%), –∏ –∑–Ω–∞—á–∏—Ç, —É –∫–∞–Ω–∞–ª–∞ –æ—á–µ–Ω—å –ª–æ—è–ª—å–Ω–∞—è –∞—É–¥–∏—Ç–æ—Ä–∏—è.

–î–æ–±–∞–≤–ª—è–µ–º –Ω–æ–≤—ã–µ –¥–∞–Ω–Ω—ã–µ:

```python
df['subscr_perc'] = df['subscriber_count'] / df['view_count'] * 100
```

```python
fig = plt.figure(constrained_layout=True, figsize=(20, 10))
sns.set(font_scale=1.5)
plt.style.use('seaborn-whitegrid')
plt.suptitle('–°–∫–æ–ª—å–∫–æ —á–µ–ª–æ–≤–µ–∫ –∏–∑ 100 –ø–æ—Å–º–æ—Ç—Ä–µ–≤—à–∏—Ö –ø–æ–¥–ø–∏—Å–∞–Ω—ã –Ω–∞ –∫–∞–Ω–∞–ª \n (–ø—Ä–æ—Ü–µ–Ω—Ç –ø—Ä–∏–≤–ª–µ—á–µ–Ω–∏—è)', 
             fontweight='bold', y=1.04, x=0.35)

spec = gridspec.GridSpec(ncols=3, nrows=2, figure=fig)

ax1 = fig.add_subplot(spec[0, :-1])
sns.scatterplot(data=df.sort_values(by='order'), x='subscr_perc', y='name', s=200, color=color)
plt.title('')
plt.xlabel('')
plt.ylabel('')

ax2 = fig.add_subplot(spec[0, -1])
sns.barplot(data=df.groupby(['name', 'order'], as_index=False).agg({'subscr_perc':'mean'})
            .sort_values(by='order'), y='name', x='subscr_perc', color=color)
plt.title('–°—Ä–µ–¥–Ω–µ–µ', fontweight='bold', y=1.02)
plt.xlabel('')
plt.ylabel('')

ax3 = fig.add_subplot(spec[1, :-1])
sns.boxplot(data=df.sort_values(by='order'), x='subscr_perc', y='name', color=color)
plt.title('')
plt.xlabel('–ø–æ–¥–ø–∏—Å–∫–∏ –æ—Ç –ø—Ä–æ—Å–º–æ—Ç—Ä–æ–≤, %')
plt.ylabel('')

ax4 = fig.add_subplot(spec[1, -1], sharex=ax2)
sns.barplot(data=df.groupby(['name', 'order'], as_index=False).agg({'subscr_perc':'median'})
            .sort_values(by='order'), y='name', x='subscr_perc', color=color)
plt.title('–ú–µ–¥–∏–∞–Ω–∞', fontweight='bold', y=1.02)
plt.xlabel('–ø–æ–¥–ø–∏—Å–∫–∏ –æ—Ç –ø—Ä–æ—Å–º–æ—Ç—Ä–æ–≤, %')
plt.ylabel('')

plt.show()
```

**–í—ã–≤–æ–¥**:

- –Ω–∞ –∫–∞–Ω–∞–ª–µ –î—ç–≤–∏–¥–∞ –ì–∞—Ä—Ä–µ—Ç—Ç–∞ –∞—É–¥–∏—Ç–æ—Ä–∏—è –æ—á–µ–Ω—å –≤–æ–≤–ª–µ—á–µ–Ω–∞. –ù–∞—à–∞ –º–µ—Ç—Ä–∏–∫–∞ –±—É–∫–≤–∞–ª—å–Ω–æ –∑–∞—à–∫–∞–ª–∏–≤–∞–µ—Ç - —Ç—É—Ç –¥–∞–∂–µ –Ω–µ 100%, –∞ –º–∞–∫—Å–∏–º—É–º 400 000% –ø—Ä–∏–≤–ª–µ—á–µ–Ω–∏—è! –ò –≤ —Å—Ä–µ–¥–Ω–µ–º —ç—Ç–æ —Ç–æ–∂–µ —Ç—ã—Å—è—á–∏ –ø—Ä–æ—Ü–µ–Ω—Ç–æ–≤. –≠—Ç–æ –∑–Ω–∞—á–∏—Ç —É –∫–∞–Ω–∞–ª–∞ –æ—á–µ–Ω—å –º–Ω–æ–≥–æ –ø–æ–¥–ø–∏—Å—á–∏–∫–æ–≤, –∏ –æ–Ω–∏ –Ω–µ –≤—Å–µ–≥–¥–∞ –ø—Ä–æ—Å–º–∞—Ç—Ä–∏–≤–∞—é—Ç –≤—ã—Ö–æ–¥—è—â–∏–µ –≤–∏–¥–µ–æ. –û–¥–Ω–∞–∫–æ —Å–≤–æ–∏ –ø–æ–¥–ø–∏—Å–∫–∏ —Å–æ—Ö—Ä–∞–Ω—è—é—Ç. –•–æ—Ä–æ—à–æ –∏–ª–∏ –ø–ª–æ—Ö–æ —ç—Ç–æ –¥–ª—è –∫–∞–Ω–∞–ª–∞? –í –¥–∞–Ω–Ω–æ–º —Å–ª—É—á–∞–µ —Å–∫–æ—Ä–µ–µ —Ö–æ—Ä–æ—à–æ - –≤–µ–¥—å –≤–æ–∑–º–æ–∂–Ω–æ –±–ª–∞–≥–æ–¥–∞—Ä—è –±–æ–ª—å—à–æ–º—É —á–∏—Å–ª—É –ø–æ–¥–ø–∏—Å—á–∏–∫–æ–≤ –î—ç–≤–∏–¥ –ì–∞—Ä—Ä–µ—Ç—Ç –∏–º–µ–µ—Ç –ª–∏–¥–µ—Ä—Å—Ç–≤–æ –ø–æ —Å—É–º–º–∞—Ä–Ω–æ–º—É —á–∏—Å–ª—É –ø—Ä–æ—Å–º–æ—Ç—Ä–æ–≤ (–∞ –≤—Å—ë-—Ç–∞–∫–∏ –ø—Ä–æ—Å–º–æ—Ç—Ä—ã - —ç—Ç–æ –ø–µ—Ä–≤–æ–æ—á–µ—Ä–µ–¥–Ω–∞—è –º–µ—Ç–∏—Ä–∏–∫–∞ –¥–ª—è –≤–∏–¥–µ–æ–ø–ª–∞—Ç—Ñ–æ—Ä–º—ã).


- –ø–æ–ª—É—á–∞–µ—Ç—Å—è, —É –Ω–∞—à–µ–π –º–µ—Ç—Ä–∏–∫–∏ –µ—Å—Ç—å –¥–≤–µ –∫—Ä–∞–π–Ω–æ—Å—Ç–∏ - —Å –æ–¥–Ω–æ–π —Å—Ç–æ—Ä–æ–Ω—ã, —É –î—ç–≤–∏–¥–∞ –ì–∞—Ä—Ä–µ—Ç—Ç–∞ –º–Ω–æ–≥–æ –ª–æ—è–ª—å–Ω—ã—Ö –ø–æ–¥–ø–∏—Å—á–∏–∫–æ–≤, –∞ –¥—Ä—É–≥–∏–µ –∫–∞–Ω–∞–ª—ã –Ω–∞ –≥—Ä–∞—Ñ–∏–∫–∞—Ö –º—ã –ø—Ä–æ—Å—Ç–æ –Ω–µ –≤–∏–¥–∏–º - –≤—Å–µ –ø–æ–∫–∞–∑–∞—Ç–µ–ª–∏ –∑–∞—Å—Ç—ã–ª–∏ –æ–∫–æ–ª–æ –Ω—É–ª—è. 

### –û–¥–æ–±—Ä–µ–Ω–∏–µ

–ù–∞ –¥–∞–Ω–Ω—ã–π –º–æ–º–µ–Ω—Ç –º—ã –≤—ã—è—Å–Ω–∏–ª–∏, —á—Ç–æ –î—ç–≤–∏–¥ –ì–∞—Ä—Ä–µ—Ç—Ç –∞–∫—Ç–∏–≤–Ω–µ–µ –≤–µ–¥—ë—Ç –∫–∞–Ω–∞–ª, –∏–º–µ–µ—Ç –ª–æ—è–ª—å–Ω—É—é –∞—É–¥–∏—Ç–æ—Ä–∏—é –∏ –ª–∏–¥–µ—Ä –ø–æ –≤—Å–µ–º —Å—É–º–º–∞—Ä–Ω—ã–º –ø–æ–∫–∞–∑–∞—Ç–µ–ª—è–º. 

–ù–æ –≤ —Å—Ä–µ–¥–Ω–µ–º –∫–∞–∂–¥–æ–µ –≤–∏–¥–µ–æ –í–∞–Ω–µ—Å—Å—ã –ú—ç–π –∏–º–µ–µ—Ç –±–û–ª—å—à–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–æ—Å–º–æ—Ç—Ä–æ–≤. 

–ó–∞–¥–∞–¥–∏–º—Å—è –≤–æ–ø—Ä–æ—Å–∞–º–∏:
- –∞ –∫–∞–∫ –∂–µ –æ—Ü–µ–Ω–∏–≤–∞–µ—Ç –ø—É–±–ª–∏–∫–∞ —ç—Ç–∏ –∫–∞–Ω–∞–ª—ã? –µ—Å–ª–∏ –±–æ–ª—å—à–µ –ø—Ä–æ—Å–º–æ—Ç—Ä–æ–≤ - –±–æ–ª—å—à–µ –∏ –ª–∞–π–∫–æ–≤? –∞ –≤ —Ç–æ–π –∂–µ —Å—Ç–µ–ø–µ–Ω–∏?

- –∫–∞–∫–æ–π –ø—Ä–æ—Ü–µ–Ω—Ç –æ—Ç –ø—Ä–æ—Å–º–æ—Ç—Ä–µ–≤—à–∏—Ö –≤–∏–¥–µ–æ —Å—Ç–∞–≤—è—Ç –ª–∞–π–∫?

- –∏ —É–≤–∏–¥–∏–º –ª–∏ –º—ã –Ω–∞ –≥—Ä–∞—Ñ–∏–∫–∞—Ö –Ω–∞—à "–º–∞–ª–µ–Ω—å–∫–∏–π" –∫–∞–Ω–∞–ª? –∏–ª–∏ —É –Ω–µ–≥–æ –≤–æ–æ–±—â–µ –Ω–µ—Ç —à–∞–Ω—Å–æ–≤? :-) 

–í—ã—á–∏—Å–ª—è–µ–º –¥–∞–Ω–Ω—ã–µ –¥–ª—è –Ω–æ–≤–æ–π –º–µ—Ç—Ä–∏–∫–∏ - –ø—Ä–æ—Ü–µ–Ω—Ç –ª–∞–π–∫–æ–≤ –æ—Ç –ø—Ä–æ—Å–º–æ—Ç—Ä–æ–≤:
