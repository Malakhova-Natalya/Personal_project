```python
class _IntArrayFormatter(pd.io.formats.format.GenericArrayFormatter):

    def _format_strings(self):
        formatter = self.formatter or (lambda x: ' {:,}'.format(x).replace(',', ' '))
        fmt_values = [formatter(x) for x in self.values]
        return fmt_values

pd.io.formats.format.IntArrayFormatter = _IntArrayFormatter
```

–†–∞—Å—Å–º–æ—Ç—Ä–∏–º –≥–ª–∞–≤–Ω—ã–µ —Ü–∏—Ñ—Ä—ã –∫–∞–Ω–∞–ª–æ–≤. –î–ª—è —ç—Ç–æ–≥–æ —Å–æ–µ–¥–∏–Ω–∏–º —Ç–∞–±–ª–∏—Ü—ã main –≤ –æ–¥–Ω—É:

```python
main = pd.concat([vdud_main, apogovorit_main, skazhi_gord_main])
main['name'] = ['–≤–î—É–¥—å*', '–ê –ø–æ–≥–æ–≤–æ—Ä–∏—Ç—å*', '–°–∫–∞–∂–∏ –ì–æ—Ä–¥–µ–µ–≤–æ–π*']
```

```python
main['view_count_mln'] = main['view_count'] / 1000000
main['subscriber_count_mln'] = main['subscriber_count'] / 1000000
```

```python
plt.figure(figsize=(20, 10))
sns.set(font_scale=1.5)
plt.style.use('seaborn-whitegrid')
plt.suptitle('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ:', fontweight='bold')

main = main.sort_values(by='view_count', ascending=False)

ax1 = plt.subplot(1, 3, 1)
sns.barplot(data=main,  x='view_count_mln', y='name', color=color)
plt.title('–ø—Ä–æ—Å–º–æ—Ç—Ä–æ–≤', fontweight='bold', y=1.02)
plt.xlabel('–º–∏–ª–ª–∏–æ–Ω—ã')
plt.ylabel('')

ax2 = plt.subplot(1, 3, 2)
sns.barplot(data=main, x='subscriber_count_mln', y='name', color=color)
plt.title('–ø–æ–¥–ø–∏—Å—á–∏–∫–æ–≤', fontweight='bold', y=1.02)
plt.xlabel('–º–∏–ª–ª–∏–æ–Ω—ã')
plt.ylabel('')

ax3 = plt.subplot(1, 3, 3)
sns.barplot(data=main, x='video_count', y='name', color=color)
plt.title('–≤–∏–¥–µ–æ', fontweight='bold', y=1.02)
plt.xlabel('–µ–¥–∏–Ω–∏—Ü—ã')
plt.ylabel('')

plt.tight_layout()
plt.show();
```
